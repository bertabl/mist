```python
hist_0 = dir.Get("_F_EnergyCH0@DT5751_1615;1")
his_rebin=hist_0.Clone()
his_rebin.Rebin(105)
print(his_rebin.GetMaximum())
print(his_rebin.GetXaxis().GetBinCenter(his_rebin.GetMaximumBin()))

c0 = ROOT.TCanvas("c0"," ",800,600)
his_rebin.Draw()
c0.Draw()
```

```python
hist_0 = dir.Get("_F_EnergyCH0@DT5751_1615;1")
hist_0.Rebin(7)
c0 = ROOT.TCanvas("c0"," ",800,600)

hist_0.Fit("gaus", "", "",1800,2600)
gfit_0= hist_0.GetFunction("gaus")
m0=gfit_0.GetParameter(1)
m0_err= gfit_0.GetParError(1)

hist_0.GetYaxis().SetTitleOffset(1.2) #1.2
hist_0.Draw()
c0.Draw()


leg0 = ROOT.TLegend(.7,.8,.9,.9)
leg0.AddEntry(hist_0,"Scintillator #1 ","L")

leg0.Draw()

mean_0 = 'Fit mean {:.1f} #pm {:.2f}'.format(m0,m0_err)
print(mean_0)

tex_fit0=ROOT.TLatex(0.5,0.7,mean_0);
tex_fit0.SetNDC();
tex_fit0.SetTextSize(0.04);
tex_fit0.Draw();

c0.SaveAs("/Users/bertabeltran/SMUT/data/Fitting_means/Scint_1.png");
```
